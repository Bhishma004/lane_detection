```python
import pickle
import numpy as np

from PIL import Image
import cv2
```

```python
train_images = pickle.load(open("data/full_CNN_train.p", "rb" ))
labels = pickle.load(open("data/full_CNN_labels.p", "rb" ))
```

```python
train_images = np.array(train_images)
labels = np.array(labels)
```

```python
train_images.shape
```

```python
train_images[1].shape
```

```python
# resize the image to the new shape
new_shape = (160,80)
resized_img = cv2.resize(train_images[1], new_shape)

# print the new shape of the image
print(resized_img.shape)

import matplotlib.pyplot as plt
plt.imshow(resized_img)
```

```python
import random
```

```python
n = 2
indices = random.sample(range(len(train_images)), n)

# Plot each image and label
fig, axs = plt.subplots(1, 2 * n, figsize=(20, 5))
for i, index in enumerate(indices):
    axs[2 * i].imshow(train_images[index])
    axs[2 * i].set_title(f'{i} Train Image')
    axs[2 * i].axis('off')
    
    axs[2 * i + 1].imshow(labels[index])
    axs[2 * i + 1].set_title(f'{i} Label Image')
    axs[2 * i + 1].axis('off')

plt.show()

n = 2
indices = random.sample(range(len(train_images)), n)

# Plot each image and label
fig, axs = plt.subplots(1, 2 * n, figsize=(20, 5))
for i, index in enumerate(indices):
    axs[2 * i].imshow(train_images[index])
    axs[2 * i].set_title(f'{i} Train Image')
    axs[2 * i].axis('off')
    
    axs[2 * i + 1].imshow(labels[index])
    axs[2 * i + 1].set_title(f'{i} Label Image')
    axs[2 * i + 1].axis('off')

plt.show()

n = 2
indices = random.sample(range(len(train_images)), n)

# Plot each image and label
fig, axs = plt.subplots(1, 2 * n, figsize=(20, 5))
for i, index in enumerate(indices):
    axs[2 * i].imshow(train_images[index])
    axs[2 * i].set_title(f'{i} Train Image')
    axs[2 * i].axis('off')
    
    axs[2 * i + 1].imshow(labels[index])
    axs[2 * i + 1].set_title(f'{i} Label Image')
    axs[2 * i + 1].axis('off')

plt.show()
```
